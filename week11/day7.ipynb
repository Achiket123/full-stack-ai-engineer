### Applying RNN, LSTM and GRU mdoels to a complete task



```python
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense, LSTM,GRU
from tensorflow.keras.models import Sequential
```

```python
VOCAB_SIZE = 10000
MAX_LEN=100
```

```python
(X_train,y_train), (X_test,y_test) = imdb.load_data(num_words=VOCAB_SIZE)

# Pad sequences
X_train = pad_sequences(X_train, maxlen=MAX_LEN, padding="post")
X_test = pad_sequences(X_test, maxlen=MAX_LEN, padding="post")

print(f"Training Data shape {X_train.shape}m {y_train.shape}")
print(f"Testing Data shape {X_test.shape}m {y_test.shape}")
```

```python
rnn_model = Sequential(
    [
        Embedding(input_dim=VOCAB_SIZE,output_dim=128)
        ,SimpleRNN(128,activation="tanh") ,
        Dense(1, activation="sigmoid"),

    ]
)
# compile
rnn_model.compile(optimizer="adam" , loss="binary_crossentropy", metrics=["accuracy"])
rnn_model.summary()
```

```python
lstm_model = Sequential(
    [
        Embedding(input_dim=VOCAB_SIZE, output_dim=128),
        LSTM(128, activation="tanh"),
        Dense(1,activation="sigmoid")
    ]
)
# compile
lstm_model.compile(optimizer="adam" , loss="binary_crossentropy", metrics=["accuracy"])
lstm_model.summary()
```

```python
gru_model = Sequential(
    [
        Embedding(input_dim=VOCAB_SIZE, output_dim=128),
        GRU(128, activation="tanh"),
        Dense(1,activation="sigmoid")
    ]
)
# compile
gru_model.compile(optimizer="adam" , loss="binary_crossentropy", metrics=["accuracy"])
gru_model.summary()
```

```python
# Train RNN Model

rnn_history = rnn_model.fit(X_train, y_train, validation_split=0.2, epochs=5, batch_size=1)
```

```python
# Train LSTM Model

lstm_history = lstm_model.fit(X_train, y_train, validation_split=0.2, epochs=5, batch_size=1)
```

```python
# Train LSTM Model

gru_history = gru_model.fit(X_train, y_train, validation_split=0.2, epochs=5, batch_size=1)
```

```python
# Evaluate Models

loss_rnn, accuracy_rnn = rnn_model.evaluate(X_test,y_test)
loss_lstm, accuracy_lstm = lstm_model.evaluate(X_test,y_test)
loss_gru, accuracy_gru = gru_model.evaluate(X_test,y_test)
```

```python
import matplotlib.pyplot as plt

plt.plot(rnn_history.history["accuracy"], label="RNN")
plt.plot(lstm_history.history["accuracy"], label="LSTM")
plt.plot(gru_history.history["accuracy"], label="GRU")

plt.title("Training accuracy comparison")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")

plt.legend()

plt.show()
```
