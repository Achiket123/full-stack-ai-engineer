### Introduction to LSTMs and how they address RNN Limitaions
*  Type of Recurrent Neural Network sepecifically designed to handle long-term dependencies
* LSTMs mitigate the vanishing gradeint problem by using specialize gate to managfe the flow of onfo
* Key Features
    * memory cells
        * maintian a long terms memory state across sequences
    * Gated Mechanism
        * Regulates how much information to keep , update or forget at each time step
    * Effective for long sequences
        * handles sequental data with dependencies across many time steps
* Advantages of LSTM over RNNs
    * retains long-term dependencies
    * Prevents gradient -related issues during training
    * Outperforms RNNs on tasks like language modeling, speech recognition and time series forecasting

### LSTM cell structure: Input , Forget and Output gates
* Forget Gate:
    * Decides what information to discard from the cell state
        * f_t = sigma( W_f . dot[h_t-1, X_t] + b_f)
            W_f: Weight matrix for the forget gate
            f_t: Forget Gate output
            X_t: input
            h_f-1 : Previous hidden states
* Input Gate
    * Decides what new information to add to the cell state
    * i_t = sigma(W_i  . [h_t-1, X_t]+ b_i)
        sigmaL sigmoid activation functions
        W_t : weight matrix for the input gate
        h_t-1: Hidden state from the previous time step
        x_t : current input
        b_i : Bias for the input gate
* cell state gate:
    * This combines the Forget gate and input gate to results the cell state
        * C_t = f_t . dot C_t-1 + i_t .dot C_t
* output gate
    * Decides what information to output at each time step
        * O_t = sigma(W_o. [h_t-1. X_t]+ b_o)
            W_o weight matrix for the output gate
            b_o Bias for the output gate

```python
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, LSTM, Dense

```

```python
VOCAB_SIZE = 10000
MAX_LEN = 200
```

```python
(X_train,y_train),(X_test, y_test) = imdb.load_data(num_words=VOCAB_SIZE)

X_train = pad_sequences(X_train, maxlen=MAX_LEN, padding="post")
X_test =  pad_sequences(X_test, maxlen=MAX_LEN, padding="post")

print(f"Training Data Shape: {X_train.shape}")
print(f"Test Data shape : {X_test.shape}")
```

```python
rnn_model = Sequential(
    [
        Embedding(input_dim=VOCAB_SIZE, output_dim=128),
        SimpleRNN(128, activation="tanh", return_sequences=False),
        Dense(1,activation="sigmoid"),
    ]
)

rnn_model.compile(optimizer="adam", loss="binary_crossentropy" , metrics=["accuracy"])
rnn_model.summary()
```

```python
rnn_history = rnn_model.fit(X_train, y_train, epochs=5, batch_size=32, validation_split=0.2)
rnn_loss, rnn_accuracy = rnn_model.evaluate(X_test,y_test)
print(f"RNN TEST LOSS: {rnn_loss} \nRNN TEST ACCURACY: {rnn_accuracy}")
```

```python
# Build and train LSTM Model
lstm_model = Sequential([
    Embedding(input_dim=VOCAB_SIZE, output_dim=128),
    LSTM(128,activation="tanh", return_sequences=False),
    Dense(1,activation="sigmoid")
])
```

```python
lstm_model.compile(optimizer="adam", loss="binary_crossentropy" , metrics=["accuracy"])
lstm_model.summary()
```

```python
lstm_history = lstm_model.fit(X_train, y_train, epochs=5, batch_size=32, validation_split=0.2)
lstm_loss, lstm_accuracy = lstm_model.evaluate(X_test,y_test)
print(f"RNN TEST LOSS: {lstm_loss} \nRNN TEST ACCURACY: {lstm_accuracy}")
```
