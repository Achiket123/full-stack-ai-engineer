```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import  ResNet50
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
```

```python
datagen = ImageDataGenerator(
    rescale= 1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    validation_split=0.2
)

train_data = datagen.flow_from_directory("./data" , target_size=(224,224), class_mode="binary", batch_size=32,subset="training")
test_data = datagen.flow_from_directory("./data" , target_size=(224,224), class_mode="binary", batch_size=32,subset="validation")
```

```python
base_model = ResNet50(weights="imagenet", include_top=False, input_shape=(224,224,4))

for layer in base_model.layers:
    layer.trainable = False

x = GlobalAveragePooling2D()(base_model.output)
output = Dense(1,activation='softmax')(x)

model = Model(inputs=base_model.input, outputs=output)
```

```python
model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
```

```python
model.fit(train_data, validation_data=test_data,epochs=10)
```

```python
val_loss, val_accuracy= model.evaluate(test_data)
print("Validation ACcuracy", val_accuracy)
```
