### Transfer Learning
* A machine learning technique where a model trained on one task is reused as a starting point for another related task
* Instead of training a model from scratch , pre-trained models are fine-tuned on a smaller dataset for a new task

* Benifits
    * Reduced Training Time
        * Pre-trained models already capture foundational features , so fewer epochs are needed
    * Improved Performance on small datasets
        * Transfer learning allows effective training even when data is limited
    * Leverages Generalization
        * Pre- trained models generalize better across tasks due to exposure to large-scaled datasets
* Applications:
    * In computer vision :
        * Pre-trained models, like ResNet, VGG, Inception , Efficient Net are used for 
            * Object detection
            * Image Classification
            * Image Segmentation
    * In NLP
        * Models like BERT , GPT , T5 are fine-tuned for 
            * Text classification
            * Sentiment Analysis
            * Named Entity recognitions
            * Questioin Answering
            


```python
import tensorflow as tf 
from tensorflow.keras.applications import ResNet50
```

```python
model = ResNet50(weights ="imagenet")
```

```python
# Display the model architecure
model.summary()
```

```python
# Access specific layer
for i, layer in enumerate(model.layers):
    print (f"Layer:{i+1}: {layer}, Trainable: {layer.trainable}")
```

```python
# freezing layers
for layer in model.layers[:-10]:
    layer.trainable = False


# Access specific layer
for i, layer in enumerate(model.layers):
    print (f"Layer:{i+1}: {layer}, Trainable: {layer.trainable}")
```
